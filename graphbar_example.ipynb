# Import the package

```python
from graphbar import graphbar as gb
```

```python
#defining the grouped data

grouped_data = {
    'Group 1':{
        'Item 1':15.5,
        'Item 2':20.3,
        'Item 3':12.5,
    },
    'Group 2':{
        'Item 1':32.5,
        'Item 2':25.3,
        'Item 3':17.5,
    },
    'Group 3':{
        'Item 1':42.5,
        'Item 2':55.3,
        'Item 3':57.5,
    },
    'Group 4':{
        'Item 1':12.5,
        'Item 2':25.9,
        'Item 3':14.5,
    }
}

#Plot the grouped data
gb.plot_bars_by_group(grouped_data=grouped_data, colors = ["bisque","tan","darkgoldenrod","darkorange"],
                     property_name="Grouped Items",ylabel="Item Value",figsize=(20,6))
```
